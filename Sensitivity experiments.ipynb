```python
#Load packages
import xarray as xr
import matplotlib.pyplot as plt 
import pandas as pd
import cartopy.crs as ccrs
```

```python
#Load topography
continents=xr.open_dataset('herold_etal_eocene_topo_1x1.nc')
#continents.topo.plot(cmap='terrain',vmin=-4500,vmax=4500)
masks=(continents.topo>0)
```

```python
#Load files 
c3_ob22_e0_p0=xr.open_dataset('c3beta_eocn_055Ma_0840ppm_022p5_E0_P0_snapshots_potsdam2-yearmean.nc')
c3_ob23_e0_p0=xr.open_dataset('c3beta_eocn_055Ma_0840ppm_023p5_E0_P0_snapshots_potsdam2_yearmean.nc')
c3_ob24_e0_p0=xr.open_dataset('c3beta_eocn_055Ma_0840ppm_024p5_E0_P0_snapshots_potsdam2-yearmean.nc')
c3_ob24_e69_p180=xr.open_dataset('c3beta_eocn_055Ma_0840ppm_024p5_E0p069_P180_snapshpts_potsdam2-yearmean.nc')
c3_ob24_e69_p270=xr.open_dataset('c3beta_eocn_055Ma_0840ppm_024p5_E0p069_P270_snapshpts_potsdam2-yearmean.nc')
c3_ob24_e69_p90=xr.open_dataset('c3beta_eocn_055Ma_0840ppm_024p5_E0p069_P90_snapshpts_potsdam2-yearmean.nc')
c3_280ppm=xr.open_dataset('c3beta_eocn_055Ma_280ppm_year22.nc')
```

```python
#PI 
yearmean=c3_280ppm['ts'].mean(dim='Time')
zonal=yearmean.mean(dim='xt_i')
```

```python
zonal.plot()
```

```python
fig,axs=plt.subplots(2,3, subplot_kw={'projection': ccrs.Robinson()},facecolor='w',figsize=(10,5)) 
diff2200=c3_ob22_e0_p0['ts'][199]-zonal
diff2200.plot(ax=axs[0,0],transform=ccrs.PlateCarree(),cmap='RdBu_r')
masks.plot.contour(ax=axs[0,0],transform=ccrs.PlateCarree(),levels=[0.5],colors='k')
axs[0,0].set_title('C3_ob22_e0_p0')

diff2300=c3_ob23_e0_p0['ts'][199]-zonal
diff2300.plot(ax=axs[0,1],transform=ccrs.PlateCarree(),cmap='RdBu_r')
masks.plot.contour(ax=axs[0,1],transform=ccrs.PlateCarree(),levels=[0.5],colors='k')
axs[0,1].set_title('c3_ob23_e0_p0')

diff2400=c3_ob24_e0_p0['ts'][199]-zonal
diff2400.plot(ax=axs[0,2],transform=ccrs.PlateCarree(),cmap='RdBu_r')
masks.plot.contour(ax=axs[0,2],transform=ccrs.PlateCarree(),levels=[0.5],colors='k')
axs[0,2].set_title('c3_ob24_e0_p0')

diff2469180=c3_ob24_e69_p180['ts'][199]-zonal
diff2469180.plot(ax=axs[1,0],transform=ccrs.PlateCarree(),cmap='RdBu_r')
masks.plot.contour(ax=axs[1,0],transform=ccrs.PlateCarree(),levels=[0.5],colors='k')
axs[1,0].set_title('c3_ob24_e69_p180')

diff2469270=c3_ob24_e69_p270['ts'][199]-zonal
diff2469270.plot(ax=axs[1,1],transform=ccrs.PlateCarree(),cmap='RdBu_r')
masks.plot.contour(ax=axs[1,1],transform=ccrs.PlateCarree(),levels=[0.5],colors='k')
axs[1,1].set_title('c3_ob24_e69_p270')

diff246990=c3_ob24_e69_p90['ts'][199]-zonal
diff246990.plot(ax=axs[1,2],transform=ccrs.PlateCarree(),cmap='RdBu_r')
masks.plot.contour(ax=axs[1,2],transform=ccrs.PlateCarree(),levels=[0.5],colors='k')
axs[1,2].set_title('c3_ob24_e69_p90')

#axs[0, 0].collections[0].colorbar.remove()

#cbar_kwargs = {'orientation': 'horizontal', 'pad': 0.05, 'aspect': 3}
#fig.colorbar(axs[0, 0].collections[0], ax=axs, **cbar_kwargs)

plt.show()
```
